# Data Cleaning

```python
mean = data['percentage_payments_gererated_within_15_days'].mean()
print(mean)
data['Remarks'] = ['Good' if x > mean else 'Bad' for x in data['percentage_payments_gererated_within_15_days']]
data["Remarks"]
```

```python
data.columns
```

```python
data1 = data.drop(['state_code','district_code','Material_and_skilled_Wages','Number_of_GPs_with_NIL_exp','Persondays_of_Central_Liability_so_far','Total_Adm_Expenditure',],axis=1)
```

Finding Outliers

```python
# numeric_cols = data1.select_dtypes(include='number').columns
numeric_cols = data1.select_dtypes(include=[np.number])
outliers_dict = {}
for col in numeric_cols:
    Q1 = np.percentile(data1[col], 25)
    Q3 = np.percentile(data1[col], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data1[(data1[col] < lower_bound) | (data1[col] > upper_bound)]
    outliers_dict[col] = len(outliers)

outliers_summary = pd.DataFrame.from_dict(outliers_dict, orient='index', columns=['Number of Outliers'])
outliers_summary.sort_values(by='Number of Outliers', ascending=False)
```

```python
top_outlier_columns = outliers_summary.head(20).index.tolist()

plt.figure(figsize=(16, 10))
for i, col in enumerate(top_outlier_columns, 1):
    plt.subplot(10,2, i)
    sns.boxplot(x=data1[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

```python
for col in numeric_cols:
    data1[col] = data1[col].astype(float)

    Q1 = np.percentile(data1[col], 25)
    Q3 = np.percentile(data1[col], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    data1[col] = np.where(data1[col] < lower_bound, lower_bound, data1[col])
    data1[col] = np.where(data1[col] > upper_bound, upper_bound, data1[col])
```

```python
top_outlier_columns = outliers_summary.head(20).index.tolist()

plt.figure(figsize=(16, 10))
for i, col in enumerate(top_outlier_columns, 1):
    plt.subplot(10,2, i)
    sns.boxplot(x=data1[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

## Distribution of Employment Among Differently Abled Persons by District 

```python
district_data = data1.groupby('district_name')['Differently_abled_persons_worked'].sum()
plt.figure(figsize=(12, 12))
plt.pie(district_data,labels=district_data.index)
plt.title('Differently Abled Persons Worked by District')
plt.show()
```

## Total Number of Completed Works Across Districts

```python
completed_works = data1.groupby('district_name')['Number_of_Completed_Works'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
completed_works.plot(kind='bar',color='blue')
plt.title('Number of Completed Works by District')
plt.xlabel('District Name')
plt.ylabel('Number of Completed Works')
plt.show()
```

## Monthly Trend of Average Employment Days Provided per Household (by Financial Month)

```python
month_order = ['April', 'May', 'June', 'July', 'August', 'September',
               'October', 'November', 'December', 'January', 'February', 'March']
data1['month'] = pd.Categorical(data1['month'], categories=month_order, ordered=True)

plt.figure(figsize=(10, 5))
for year in data1['fin_year'].unique():
    yearly_data = data1[data1['fin_year'] == year]
    avg_by_month = yearly_data.groupby('month',observed=False)['Average_days_of_employment_provided_per_Household'].mean().reindex(month_order)
    plt.plot(month_order, avg_by_month, marker='o', label=year)

plt.title('Avg Employment Days per Household (by Month)')
plt.xlabel('Month')
plt.ylabel('Avg Days')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## Monthly Trend of Total Individuals Worked

```python
# 2. Line Chart: Total Individuals Worked per Month
plt.figure(figsize=(10, 5))
for year in data1['fin_year'].unique():
    yearly_data = data1[data1['fin_year'] == year]
    total_by_month = yearly_data.groupby('month',observed=False)['Total_Individuals_Worked'].sum().reindex(month_order)
    plt.plot(month_order, total_by_month, marker='s', label=year)

plt.title('Total Individuals Worked (by Month)')
plt.xlabel('Month')
plt.ylabel('Total Individuals')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## Total Number of Households Worked by District

```python
columns_of_interest = [
    'Approved_Labour_Budget',
    'Average_Wage_rate_per_day_per_person',
    'Average_days_of_employment_provided_per_Household',
    'Differently_abled_persons_worked',
    'Total_No_of_JobCards_issued',
    'Total_No_of_Workers',
    'Total_No_of_Works_Takenup',
    'Wages',
    'Women_Persondays',
    'percent_of_Category_B_Works',
    'percent_of_Expenditure_on_Agriculture_Allied_Works',
    'percent_of_NRM_Expenditure',
    'percentage_payments_gererated_within_15_days'
]

df_cleaned = data1[columns_of_interest].apply(pd.to_numeric, errors='coerce')

correlation_matrix = df_cleaned.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Employment Metrics Correlation - Assam")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
