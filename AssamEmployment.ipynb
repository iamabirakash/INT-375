```python
data1 = data.drop(['state_code','district_code','Material_and_skilled_Wages','Number_of_GPs_with_NIL_exp','Persondays_of_Central_Liability_so_far','Total_Adm_Expenditure',],axis=1)
```

```python
# Identify numerical columns
# numeric_cols = data1.select_dtypes(include='number').columns
numeric_cols = data1.select_dtypes(include=[np.number])
# Dictionary to store outliers for each numeric column
outliers_dict = {}

# Use IQR method to find outliers in each numeric column
for col in numeric_cols:
    Q1 = np.percentile(data1[col], 25)
    Q3 = np.percentile(data1[col], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data1[(data1[col] < lower_bound) | (data1[col] > upper_bound)]
    outliers_dict[col] = len(outliers)

outliers_summary = pd.DataFrame.from_dict(outliers_dict, orient='index', columns=['Number of Outliers'])
outliers_summary.sort_values(by='Number of Outliers', ascending=False)
```

```python
top_outlier_columns = outliers_summary.head(20).index.tolist()

plt.figure(figsize=(16, 10))
for i, col in enumerate(top_outlier_columns, 1):
    plt.subplot(10,2, i)
    sns.boxplot(x=data1[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

```python
# Capping outliers safely by converting columns to float
for col in numeric_cols:
    # Convert column to float to avoid dtype issues
    data1[col] = data1[col].astype(float)

    Q1 = np.percentile(data1[col], 25)
    Q3 = np.percentile(data1[col], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Apply capping
    data1[col] = np.where(data1[col] < lower_bound, lower_bound, data1[col])
    data1[col] = np.where(data1[col] > upper_bound, upper_bound, data1[col])
```

```python
top_outlier_columns = outliers_summary.head(20).index.tolist()

plt.figure(figsize=(16, 10))
for i, col in enumerate(top_outlier_columns, 1):
    plt.subplot(10,2, i)
    sns.boxplot(x=data1[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

```python
month_order = ['April', 'May', 'June', 'July', 'August', 'September',
               'October', 'November', 'December', 'January', 'February', 'March']
data1['month'] = pd.Categorical(data1['month'], categories=month_order, ordered=True)

# 1. Line Chart: Avg Employment Days per Month
plt.figure(figsize=(10, 5))
for year in data1['fin_year'].unique():
    yearly_data = data1[data1['fin_year'] == year]
    avg_by_month = yearly_data.groupby('month',observed=False)['Average_days_of_employment_provided_per_Household'].mean().reindex(month_order)
    plt.plot(month_order, avg_by_month, marker='o', label=year)

plt.title('Avg Employment Days per Household (by Month)')
plt.xlabel('Month')
plt.ylabel('Avg Days')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 2. Line Chart: Total Individuals Worked per Month
plt.figure(figsize=(10, 5))
for year in data1['fin_year'].unique():
    yearly_data = data1[data1['fin_year'] == year]
    total_by_month = yearly_data.groupby('month',observed=False)['Total_Individuals_Worked'].sum().reindex(month_order)
    plt.plot(month_order, total_by_month, marker='s', label=year)

plt.title('Total Individuals Worked (by Month)')
plt.xlabel('Month')
plt.ylabel('Total Individuals')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 3. Heatmap: Avg Employment Days (Month vs Year)
heatmap_data = data1.pivot_table(index='month', columns='fin_year', observed=False,
                              values='Average_days_of_employment_provided_per_Household',
                              aggfunc='mean').reindex(month_order)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt=".1f", cbar=True)
plt.title('Heatmap of Avg Employment Days')
plt.xlabel('Financial Year')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```

```python
# 1. Bar chart: Total Households Worked by District
plt.figure(figsize=(14, 6))
district_work = data1.groupby('district_name',observed=False)['Total_Households_Worked'].sum().sort_values(ascending=False)
sns.barplot(x=district_work.values, y=district_work.index,hue=district_work.index, palette='viridis')
plt.title('Total Households Worked by District')
plt.xlabel('Total Households Worked')
plt.ylabel('District')
plt.tight_layout()
plt.show()
```

```python
# 2. Pie chart: Top 10 Districts by Total No. of Job Cards Issued
top10_jobcards = data1.groupby('district_name')['Total_No_of_JobCards_issued'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 8))
plt.pie(top10_jobcards, labels=top10_jobcards.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
plt.title('Top 10 Districts by Job Cards Issued')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
